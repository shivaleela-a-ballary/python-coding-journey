```python
'''
1. Fizz Buzz 
Question

Write a program that accepts an integer n and returns a list of strings from 1 to n.

Follow these rules:

If a number is divisible by 3, print "Fizz".
If a number is divisible by 5, print "Buzz".
If a number is divisible by both 3 and 5, print "FizzBuzz".
Otherwise, print the number itself.
'''
#Solution
def fizzBuzz(n):

    result = []

    for i in range(1, n + 1):

        if i % 15 == 0:
            result.append("FizzBuzz")

        elif i % 3 == 0:
            result.append("Fizz")

        elif i % 5 == 0:
            result.append("Buzz")

        else:
            result.append(str(i))

    return result

print(fizzBuzz(15))
```

```python
'''
2. Number of Steps to Reduce a Number to Zero (Easy)
Question

Given an integer num, reduce it to zero using the following rules:

If the number is even, divide it by 2.
If the number is odd, subtract 1.

Return the total number of steps required to reach 0.
'''
#Solution

def numberOfSteps(num):

    steps = 0

    while num != 0:

        if num % 2 == 0:
            num //= 2

        else:
            num -= 1

        steps += 1

    return steps

print(numberOfSteps(14))
```

```python
'''
3. Palindrome Number (Easy)
Question

Given an integer x, determine whether it reads the same forward and backward.

Return:

True if it is a palindrome.
False otherwise.
'''
#Solution
def isPalindrome(x):

    if x < 0:
        return False

    original = x
    reverse = 0

    while x > 0:

        digit = x % 10
        reverse = reverse * 10 + digit
        x //= 10

    return original == reverse

print(isPalindrome(121))
print(isPalindrome(123))
```

```python
'''
4.Running Sum of 1D Array (Easy)
Question

Given an array of integers, create a new array where each element is the sum of all previous elements including the current element.
'''
#Solution
def runningSum(nums):

    for i in range(1, len(nums)):
        nums[i] += nums[i-1]

    return nums

print(runningSum([1,2,3,4]))
```

```python
'''
5.Richest Customer Wealth (Easy)
Question

Each row of a 2D array represents the bank accounts of one customer.

Find and return the maximum total wealth among all customers.
'''
#Solution
def maximumWealth(accounts):

    maxWealth = 0

    for customer in accounts:

        wealth = sum(customer)

        if wealth > maxWealth:
            maxWealth = wealth

    return maxWealth

accounts = [
    [1,2,3],
    [3,2,1]
]

print(maximumWealth(accounts))
```
